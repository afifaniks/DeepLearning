```python
torch.manual_seed(2)

random_weight = torch.randn((2, 20))

random_like = torch.randn_like(random_weight)

print(random_weight, "\n",  random_like)

torch.randn((1, 1))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

x = random_weight.numpy()
print(random_weight)
print(torch.mean(random_weight))
print(torch.median(random_weight))
plt.plot(x)
plt.show()
```

```python
# Normal Distribution Check

random = torch.randn((1, 10000))

smaller = 0
greater = 0
x = random.numpy()

mean = torch.mean(random).numpy()

print("Mean ", mean)

for i in range(10000):
    if x[0][i] < mean:
        smaller += 1
    else:
        greater +=1

print("Smaller ", smaller, "\nGreater", greater)
```
